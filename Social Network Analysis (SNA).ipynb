##### undirected graph

```python
## create a object of Graph how we create array = array[] then we will add items to it
g = nx.Graph()
## to add one node to the graph use add_nodes to add multiple nodes use add_nodes_from
g.add_node(1)  ## to add node name 1 
## g.add_node(1)
## g.remove_node(1)
g.add_nodes_from([1,2,3,'five'])
g.nodes()
## note there is a difference between Graph and graph Graph is a class in the networkx module where graph is a atrtribute
```

```python
% matplotlib inline 
nx.draw(g,with_labels=True)
```

```python
g2 = nx.Graph()
g2.add_edges_from([('A','B'),('B','D'),('D','C'),('D','E'),('D','F')])
nx.draw(g2,with_labels=True)
```

```python
### UNDIRECTED GRAPH
g3 = nx.DiGraph()
g3.add_edges_from([('A','B'),('B','D'),('D','C'),('D','E'),('D','F')])
## g3.add_edges_from(g2)
nx.draw(g3,with_labels=True)
```
