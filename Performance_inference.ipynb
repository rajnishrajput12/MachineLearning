```python
def get_quantile(datafram,quantile):
    df = pd.DataFrame([]) ## ceating empty dataframe
    for i in datafram['job_dsc'].unique():
        for j in datafram[datafram['job_dsc']==i]['experience_bin'].unique():
            temp = datafram[(datafram['job_dsc']==i)& (datafram['experience_bin']==j)]
            x =temp['bonus_ratio'].quantile(quantile)
            y = temp['salary_hike'].quantile(quantile)
            z = temp['compa_rt'].quantile(quantile)
            temp['Bonus_quantile'] = x
            temp['SalaryHike_quantile'] = y
            temp['CompaRT_quantile'] = z
            df = df.append(temp,ignore_index =True)
    return df   
```

```python
def fillna(dataframe):
    dataframe.fillna(0,inplace=True)
```

```python
'''
this function will fill bonus column with the median 
'''
def clean_data(datafram):
    print(datafram.isnull().sum())
    print('replacing the null values to the mean of its group we will drop if its doesnt find any value in its group')
    datafram['bonus_ratio'] = datafram.groupby(['job_dsc','year','experience_bin']).transform(lambda x: x.fillna(x.median()))
    datafram.dropna(inplace=True)
    return datafram
```

```python
'''
this will plot the histogram of feature in a datarame for a given year
'''
def plot_feature(feature,dataframe,year,all=None):
    bins=[]
    print(all)
    if(all==None):
        dataframe = dataframe[dataframe['year']==year]
        print('here')
    else:        
        dataframe = dataframe[dataframe['job_dsc'].isin(dataframe[dataframe['year']==year].groupby(['job_dsc']).aoid.nunique().nlargest(10).index)]
    if(feature=='salary_hike'):
        bins=np.arange(0,60,5)
    elif(feature=='bonus_ratio'):
        bins=np.arange(0,3,0.2)
    elif(feature=='compa_rt'):
        bins=np.arange(0,3,0.2)
    ax = dataframe[dataframe['year']==year][feature].hist(bins=bins,figsize=(20,10))
    ax.set_xlabel(feature)
    ax.set_ylabel("FREQUENCY")
    ax.set_title( feature+' Distribution for top 10 job',{'fontsize': 20, 'fontweight': 'medium'})
```

```python
'''
this function will plot a box plot for top 10 job
'''
def get_box_plot(dataframe,feature,year):
    plt.figure(figsize=(20,12))
    dataframe = dataframe[dataframe['job_dsc'].isin(dataframe[dataframe['year']==year].groupby(['job_dsc']).aoid.nunique().nlargest(10).index)]
    sns.boxplot(x="job_dsc", y=feature, data= dataframe,sym="")
    # Changing default seaborn/matplotlib to be more readable
    plt.xlabel('job_dsc', fontsize = 15)
    plt.ylabel(feature, fontsize = 15)
    plt.title(feature+" Boxplot for top 10 job")
    plt.xticks(fontsize = 12,rotation=90)
    plt.yticks(fontsize = 12)
    if(feature=='compa_rt'):
        plt.ylim(np.arange(0,2))
```

```python
def get_quantile(data,year,Bonus_quant,SalaryHike_quant,CompaRT_quant):
    g1 = data[data['year']==year].groupby(['job_dsc','experience_bin']).bonus_ratio.quantile(Bonus_quant)
    g1 = g1.to_frame(name='Bonus_quantile').reset_index()
    df_qntile = pd.merge(data[data['year']==year],g1[['job_dsc','experience_bin','Bonus_quantile']],on=['job_dsc','experience_bin'])
    g2 = data[data['year']==year].groupby(['job_dsc','experience_bin']).salary_hike.quantile(SalaryHike_quant)
    g2 = g2.to_frame(name='SalaryHike_quantile').reset_index()
    df_qntile = pd.merge(df_qntile,g2[['job_dsc','experience_bin','SalaryHike_quantile']],on=['job_dsc','experience_bin'])
    g3 = data[data['year']==year].groupby(['job_dsc','experience_bin']).compa_rt.quantile(CompaRT_quant)
    g3 = g3.to_frame(name='CompaRT_quantile').reset_index()
    df_qntile = pd.merge(df_qntile,g3[['job_dsc','experience_bin','CompaRT_quantile']],on=['job_dsc','experience_bin'])
    del(g1,g2,g3)
    return df_qntile
    del(df_qntile)
```

```python
def get_data_Detail(data,year):
    b1 = data[data['year']==year].groupby(['job_dsc','experience_bin']).bonus_ratio.quantile(0.25)
    b1 = b1.to_frame(name='bonus_25_quantile').reset_index()
    b2 = data[data['year']==year].groupby(['job_dsc','experience_bin']).bonus_ratio.quantile(0.75)
    b2 = b2.to_frame(name='bonus_75_quantile').reset_index()
    b2  = pd.merge(b1,b2,on=['job_dsc','experience_bin'])
    s1 = data[data['year']==year].groupby(['job_dsc','experience_bin']).salary_hike.quantile(0.75)
    s1 = s1.to_frame(name='SalaryHike_75_quantile').reset_index()
    s2 = data[data['year']==year].groupby(['job_dsc','experience_bin']).salary_hike.quantile(0.25)
    s2 = s2.to_frame(name='SalaryHike_25_quantile').reset_index()
    s2 = pd.merge(s1,s2,on=['job_dsc','experience_bin'])
    dataframe_table = pd.merge(b2,s2,on=['job_dsc','experience_bin'])
    c1 = data[data['year']==year].groupby(['job_dsc','experience_bin']).compa_rt.quantile(0.25)
    c1 = c1.to_frame(name='compart_25_quantile').reset_index()
    c2 = data[data['year']==year].groupby(['job_dsc','experience_bin']).compa_rt.quantile(0.75)
    c2 = c2.to_frame(name='compart_75_quantile').reset_index()
    c2  = pd.merge(c1,c2,on=['job_dsc','experience_bin'])
    dataframe_table = pd.merge(dataframe_table,c2,on=['job_dsc','experience_bin'])
    h1 = data[data['year']==year].groupby(['job_dsc','experience_bin']).aoid.nunique()
    h1 = h1.to_frame(name='no_of_emp').reset_index()
    dataframe_table = pd.merge(dataframe_table,h1,on=['job_dsc','experience_bin'])
    bm1 = data[data['year']==year].groupby(['job_dsc','experience_bin']).bonus_ratio.agg(['mean','median','std'])
    bm1 = bm1.reset_index()
    bm1.rename(columns = {'mean':'Bonus_Mean','median':'Bonus_median','std':'Bonus_std'}, inplace = True)
    dataframe_table = pd.merge(dataframe_table,bm1,on=['job_dsc','experience_bin'])
    sm1 = data[data['year']==year].groupby(['job_dsc','experience_bin']).salary_hike.agg(['mean','median','std'])
    sm1 = sm1.reset_index()
    sm1.rename(columns = {'mean':'Salary_Mean','median':'Salary_median','std':'Salary_std'}, inplace = True)
    dataframe_table = pd.merge(dataframe_table,sm1,on=['job_dsc','experience_bin'])
    cm1 = data[data['year']==year].groupby(['job_dsc','experience_bin']).compa_rt.agg(['mean','median','std'])
    cm1 = cm1.reset_index()
    cm1.rename(columns = {'mean':'Compart_Mean','median':'Compart_median','std':'Compart_std'}, inplace = True)
    dataframe_table = pd.merge(dataframe_table,cm1,on=['job_dsc','experience_bin'])
    dataframe_table = dataframe_table[['job_dsc','experience_bin','no_of_emp','Bonus_Mean','Bonus_median','Bonus_std','bonus_25_quantile','bonus_75_quantile','Salary_Mean','Salary_median','Salary_std','SalaryHike_25_quantile','SalaryHike_75_quantile','Compart_Mean','Compart_median','Compart_std','compart_25_quantile','compart_75_quantile']]
    dataframe_table.fillna(0,inplace=True)
    dataframe_table.sort_values('no_of_emp',ascending=False,inplace=True)
    return dataframe_table
    del(b1,b2,s1,s2,c1,c2,h1,bm1,sm1,cm1,dataframe_table)
```
