```python
def cal_grad(x,y):
    learn_rate =0.001
    m_curr = b_curr =0
    n = len(x)
    for i in range(1000):
        y_pred = (m_curr*x)+b_curr
        error = 1/n*(sum([val**2 for val in (y-y_pred)]))
        dm = -2/n*(sum(y-y_pred))
        db = -2/n*(sum(y-y_pred))
        m_curr = m_curr - learn_rate*dm
        b_curr = b_curr- learn_rate*db        
        print("m:-"+str(m_curr)+"b:-"+str(b_curr)+"error:"+str(error)+"iteraiton:-"+str(i))
```

```python
cal_grad(x,y)
```
