from this point to K-number of neighbours

Step 1:-choose K and decide how many number of neighbout we will be participating in voting
    2:- calcualte the eucledian distance of the given point from all the point in the dataset
    3.:- now based on K we will take these many distance in our voting
    4.:- based how the majority of the class out of all K class the given data will belong to that class
