```python
import pandas as pd
## dataframe from list 
df1 = pd.DataFrame({'City': ["Patna","Delhi","kochin"],
                    'Temperature' : ["100","20","90"]
                   })

df2 = pd.DataFrame({'City' : ["Patna","Delhi","kochin","Hyderabad"],
                    'Population_in_Millions' : ["12","40","20","22"]
    
})  
df2
## merge data frame 
# merge will merge two df on given condtion(common colum) it is almost like join in SQL
## https://www.youtube.com/watch?v=h4hOPGo4UVU&list=PLeo1K3hjS3uuASpe-1LjfG5f14Bnozjwy&index=9
df_merge = pd.merge(df1,df2, on= "City", how="outer")
print(df_merge)
```
